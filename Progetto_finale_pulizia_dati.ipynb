```python
imdb_movies_clean = imdb_movies.copy()
```

#### Ma questo df ha un problema: i film con lo stesso titolo non sono facilmente distinguibili, se non confrontando la data di uscita

```python
imdb_movies_clean.loc[imdb_movies_clean['names'] == 'Pinocchio']
```

Ad esempio, se cerchiamo il film "Pinocchio" vediamo che esistono varie versioni di anni, genere e lingua diversi; e che in questi casi i dati di budget e fatturato non sono attendibili.


#### Occorre anzitutto identificare quali sono i duplicati e cosa è duplicato

```python
# identifichiamo i duplicati
doppioni = imdb_movies_clean[imdb_movies_clean.duplicated(subset='names', keep=False)]
doppioni.sort_values(by=['names'])
```

- In alcuni casi si nota che i valori di budget e fatturato sono stati copiati per film omonimi (ad esempio "12 Angry Men")
- In altri casi lo stesso film ha due righe con valori monetari diversi (come "20,000 Leagues Under the Sea"). In entrambi i casi non abbiamo una entry attendibile.

In un primo passaggio identifico le righe che effettivamente si ripetono, controllando sia il nome, la data di uscita che la revenue.

```python
doppioni = imdb_movies_clean[imdb_movies_clean.duplicated(subset=['names','date_x','revenue'], keep=False)]
doppioni
```

In questo caso non ottengo nulla. Significa che non ci sono righe ripetute, quindi tutti i duplicati di nome e data avranno diversi fatturati.

#### Procedo allora eliminando **tutti** i film con lo stesso titolo e data, poiché non ho modo di verificare agilmente quali siano i valori di fatturato corretti quando ne ho diversi per lo stesso film.

```python
doppioni = imdb_movies_clean[imdb_movies_clean.duplicated(subset=['names','date_x'], keep=False)]
doppioni
```

```python
# eliminazione di tutti i duplicati
indice_duplicati = doppioni.index
imdb_movies_clean = imdb_movies_clean.drop(indice_duplicati)
imdb_movies_clean.shape
# 10,178 righe originali - 347 righe dei doppioni = 9831 righe finali
```

#### Verifico se esistono ancora degli omonimi, e che abbiano dei valori diversi pur avendo lo stesso titolo

```python
omonimi = imdb_movies_clean['names'].value_counts()
omonimi = omonimi[omonimi > 1]
omonimi
```

```python
imdb_movies_clean[imdb_movies_clean['names'] == 'Jesus']
```

#### Ora possiamo passare alla prossima fase di pulizia: tenere solo le colonne interessanti e rinominarle

```python
# sistemazione colonne
imdb_movies_clean = imdb_movies_clean.drop(['status'], axis=1)
imdb_movies_clean = imdb_movies_clean.rename(columns={'names':'movie_title','date_x':'release_date','score':'imdb_score'})
imdb_movies_clean.head(1)
```

```python
imdb_movies_clean['release_year'] = pd.to_datetime(imdb_movies_clean['release_date']).dt.year # estraiamo l'anno di uscita
```

```python
imdb_movies_clean.info()
```

```python
#salviamo il csv
imdb_movies_clean.to_csv("C:\\Users\\loren\\Desktop\\Final Project\\imdb_movies_clean", index=False)
```

#### Controllo duplicati

```python
# controlliamo i duplicati di nome & anno di uscita & nome regista
doppioni = imdb5000_clean[imdb5000_clean.duplicated(subset=['movie_title','title_year','director_name'], keep=False)]
doppioni.sort_values('movie_title')
```

Avendo controllato per 3 colonne, possiamo essere sicuri che questi siano duplicati, dunque li eliminiamo tenendo solo il primo

```python
imdb5000_clean = imdb5000_clean.drop_duplicates(subset = ['movie_title','title_year','director_name'], keep = 'first')
```

#### Droppiamo i film senza anno di uscita

```python
no_anno = imdb5000_clean.loc[imdb5000_clean['title_year'].isna()].index
imdb5000_clean = imdb5000_clean.drop(index=no_anno)
```

## Merge

#### Uniamo i dati del secondo df a quelli del primo, facendo una merge su due chiavi: titolo e anno di uscita

Facciamo anzitutto una normalizzazione delle colonne titolo: tutto in minuscolo e rimozione di spazi vuoti e caratteri

```python
imdb_movies_clean['movie_title'] = imdb_movies_clean['movie_title'].str.strip().str.lower() # normalizzazione del titolo
imdb5000_clean['movie_title'] = imdb5000_clean['movie_title'].str.strip().str.lower()       # per entrambi i df
```

```python
df_movies_merged = pd.merge(
    imdb_movies_clean,                      # df da 10k righe
    imdb5000_clean,                         # df secondario
    left_on=['movie_title','release_year'], # doppia chiave per evitare gli omonimi
    right_on=['movie_title','title_year'],
    how='inner'                             # inner join per tenere solo le righe con tutti i dati      
)
```

```python
df_movies_merged
```

## Pulizia finale

```python
df_movies_merged.head(1)
```

```python
df_movies_merged.info()
```

#### Gestione delle colonne

```python
df_movies_merged = df_movies_merged.drop(['title_year','crew'], axis=1)
```

```python
df_movies_merged = df_movies_merged.rename(columns={'duration':'duration[min]','budget_x':'budget', 'num_critic_for_reviews':'review_num'})
df_movies_merged.head(1)
```

#### Trasformazione dei valori float non necessari in int (`imdb_score`, `budget`, `revenue`, `duration`, `review_num`, `director_facebook_likes`)

```python
colonne = ['imdb_score','budget','revenue','duration[min]','review_num','director_facebook_likes']
df_movies_merged[colonne] = df_movies_merged[colonne].round(0).astype('Int64') # arrotondo e poi cambio type
```

#### Creazione della colonna `cast` unendo i 3 attori/attrici principali

Al momento la colonna del cast è divisa in 3 colonne diverse, uniamole in una unica che chiameremo `cast`

```python
df_movies_merged['cast'] = df_movies_merged['actor_1_name'] + ', ' + df_movies_merged['actor_2_name'] + ', ' + df_movies_merged['actor_3_name']

```

Cancelliamo le colonne divise

```python
df_movies_merged = df_movies_merged.drop(['actor_1_name','actor_2_name','actor_3_name'], axis=1)
```
